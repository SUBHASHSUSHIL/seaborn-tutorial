# FacetGrid

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
fg = sns.FacetGrid(var, col="time", row="smoker", margin_titles=True, height=4, aspect=1.5)
fg.map(sns.scatterplot, "total_bill", "tip")
plt.show()
```
