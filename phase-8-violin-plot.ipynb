# Violin Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day", y="total_bill", data=var)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=var, hue="time")
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=var, order=["Dinner", "Lunch"])
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=var, hue="sex", split=True, density_norm="width")
plt.show()
```
