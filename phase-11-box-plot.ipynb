# Box Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")

sns.boxenplot(x="day", y="total_bill", data=var, hue="sex", color="navy", palette="pastel", order=["Thur", "Fri", "Sat", "Sun"])
plt.show()
```
