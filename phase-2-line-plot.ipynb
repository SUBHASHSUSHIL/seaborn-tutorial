## Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = [1, 2, 3, 4, 5, 6, 7]
var_1 = [2, 3, 4, 1, 5, 6, 7]

data = pd.DataFrame({'var': var, 'var_1': var_1})
sns.lineplot(data=data, x='var', y='var_1')

plt.show()
```

```python

plt.plot(var, var_1)
plt.show()
```

```python
y_1 = sns.load_dataset("penguins")

y_1
```

```python
sns.lineplot(data=y_1, x="bill_depth_mm", y="flipper_length_mm", hue="sex")
plt.show()
```

```python
sns.lineplot(data=y_1, x="bill_depth_mm", y="flipper_length_mm", hue="sex", style="species", markers=True)
plt.show()
```
