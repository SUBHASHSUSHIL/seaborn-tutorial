# Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.scatterplot(data=var, x="flipper_length_mm", y="bill_length_mm", hue="species", style="sex", size="body_mass_g", sizes=(20, 200))
plt.show()
```

```python
sns.scatterplot(data=var, x="flipper_length_mm", y="bill_length_mm", hue="species")
plt.show()
```
