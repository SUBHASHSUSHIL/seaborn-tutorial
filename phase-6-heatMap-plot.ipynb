# HeatMap Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var = np.linspace(0, 10, 20).reshape(4, 5)
var
```

```python
sns.heatmap(var, annot=True, vmax=12)
plt.show()
```

```python
data = sns.load_dataset("anagrams")
x = data.drop(columns=["attnr"], axis=1).head(10)
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(var, vmin=0, annot=True, vmax=12, cmap="PuOr")
plt.show()
```
