# Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.barplot(x="species", y="flipper_length_mm", data=var)
plt.show()
```

```python
sns.barplot(x="species", y="flipper_length_mm", data=var, hue="sex")
plt.show()
```

```python
order_species = ["Gentoo", "Adelie", "Chinstrap"]
sns.barplot(x="species", y="flipper_length_mm", data=var, hue="sex", order=order_species)
plt.show()
```

```python
order_species = ["Gentoo", "Adelie", "Chinstrap"]
sns.barplot(x="species", y="flipper_length_mm", data=var, hue="sex", order=order_species, hue_order=["Female", "Male"], palette="Set2")
plt.show()
```

```python
order_species = ["Gentoo", "Adelie", "Chinstrap"]
sns.barplot(x="species", y="flipper_length_mm", data=var, hue="sex", order=order_species, hue_order=["Female", "Male"],
            saturation=0.1)
plt.show()
```

```python
order_species = ["Gentoo", "Adelie", "Chinstrap"]
sns.barplot(x="species", y="flipper_length_mm", data=var, hue="sex", order=order_species, hue_order=["Female", "Male"],
            saturation=0.1, capsize=0.2)
plt.show()
```
