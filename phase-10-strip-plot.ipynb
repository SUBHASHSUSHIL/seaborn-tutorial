# Strip Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.stripplot(var)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, jitter=True, hue="sex", dodge=True, marker="D", size=7, edgecolor="gray", alpha=0.7)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="rocket_r", linewidth=1)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="rocket_r", linewidth=1, edgecolor="r", jitter=5)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="rocket_r", linewidth=1, edgecolor="r", size=5, marker="<")
plt.show()
```
