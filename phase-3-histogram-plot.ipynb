# Histogram Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.histplot(data=var, x="flipper_length_mm", kde=True)
plt.show()
```

```python
sns.histplot(data=var, x="flipper_length_mm", kde=True, bins= [170, 180, 190, 200, 210, 220, 230, 240])
plt.show()
```

```python
sns.histplot(data=var, x="flipper_length_mm", kde=True, color='green', log_scale=True)
plt.show()
```
